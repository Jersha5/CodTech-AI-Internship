```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
prompt = input("Enter a topic: ")

output = generator(
    prompt,
    max_length=120,
    num_return_sequences=1,
    temperature=0.8
)

print("\nGenerated Text:\n")
print(output[0]["generated_text"])
```

```python
while True:
    prompt = input("\nEnter a topic (or type 'exit' to quit): ")

    if prompt.lower() == "exit":
        print("Thank you for using the Text Generation Model!")
        break

    output = generator(
        prompt,
        max_length=120,
        num_return_sequences=1,
        temperature=0.8
    )

    print("\nGenerated Text:\n")
    print(output[0]["generated_text"])
```

```python
while True:
    prompt = input("\nEnter a topic (Type 'exit' to quit): ")

    # Check if the input is empty
    if prompt.strip() == "":
        print("Please enter a valid topic.")
        continue

    # Exit the program
    if prompt.lower() == "exit":
        print("Thank you for using the Generative Text Model!")
        break

    # Generate text
    output = generator(
        prompt,
        max_length=120,
        num_return_sequences=1,
        temperature=0.8,
        do_sample=True
    )

    print("\nGenerated Text:\n")
    print(output[0]["generated_text"])
```
