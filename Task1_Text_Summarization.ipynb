```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python
text = """
Artificial Intelligence (AI) is transforming industries around the world.
It enables computers to perform tasks that normally require human intelligence,
such as speech recognition, language understanding, image recognition,
and decision making.

AI is widely used in healthcare, education, banking,
transportation, and manufacturing.

Doctors use AI to detect diseases early.
Banks use AI to identify fraudulent transactions.
Self-driving cars use AI for navigation.

Although AI offers many benefits,
it also raises concerns regarding privacy,
job displacement, and ethical decision making.

Researchers continue improving AI technologies
to make them more reliable, safe, and beneficial for society.
"""
```

```python
parser = PlaintextParser.from_string(text, Tokenizer("english"))

summarizer = LsaSummarizer()

summary = summarizer(parser.document, 3)

print("========== ORIGINAL TEXT ==========\n")
print(text)

print("\n========== SUMMARY ==========\n")

for sentence in summary:
    print(sentence)
```
